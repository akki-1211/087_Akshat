```python
# import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read the data
data = pd.read_csv('Data_for_Transformation.csv')
print(data)
```

```python
# the scatter plot between age and salary
plt.scatter(data['Age'], data['Salary'])
plt.show()
```

```python
# draw histogram of salary
plt.hist(data['Salary'], bins=5, color='red')
plt.show()
```

```python
# draw histogram of salary
plt.hist(data['Salary'], bins=10, color='purple')
plt.show()
```

```python
# draw histogram of salary
plt.hist(data['Salary'], bins=15, color='lightblue')
plt.show()
```

```python
# Draw bar chart of country
plt.bar(data['Country'], data['Salary'], color='lightblue')
plt.title("Bar graph of Country vs Salary", fontweight="bold")
plt.xlabel("Country")
plt.ylabel("Salary")
plt.show()
```
